```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="white")


# Compute the correlation matrix
corr = metro_df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
corr_mat = sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=False, linewidths=.5)

plt.title("Correlation Matrix", fontsize=25)

corr_mat.figure.savefig("Visualizations\CorrMat2.png")
```

```python
features = list(metro_df.columns)
```

```python
features.remove('City')
features.remove("Country")
features.remove("Annual ridership")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["feature"] = features
vif["VIF"] = [variance_inflation_factor(metro_df[features].values, i) for i in range(len(features))]
```
