```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="white")


# Compute the correlation matrix
corr = metro_df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
corr_mat = sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=False, linewidths=.5)

plt.title("Correlation Matrix", fontsize=25)

corr_mat.figure.savefig("Visualizations\CorrMat2.png")
```

```python
# Create a figure with 4x4 subplots
fig, axs = plt.subplots(4, 4, figsize=(10, 10))

# Loop through every column in the DataFrame
for i, col in enumerate(metro_df_bw.columns):
    # Create a box and whisker plot for the current column
    axs[i//4, i%4].boxplot(metro_df_bw[col])
    axs[i//4, i%4].set_title(col)

plt.title("Distributions of Variables")

# Adjust the layout of the subplots
plt.tight_layout()



# Display the plot
plt.show()
```
