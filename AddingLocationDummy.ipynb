```python
import pandas as pd

preprocessed_df = pd.read_csv('data/final_data.csv')
```

```python
preprocessed_df
```

```python
import pycountry_convert as pc

# create a function to map a country to its continent using pycountry_convert
def get_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return country_name
```

```python
preprocessed_df['Continent'] = preprocessed_df['Country'].apply(get_continent)
```

```python
import numpy as np
list_consts = np.unique(list(preprocessed_df['Continent']))

for cont in list_consts:
    preprocessed_df[cont] = preprocessed_df.apply((lambda row: 1 if row['Continent'] == cont else 0), axis=1)
```

```python
preprocessed_df.to_csv('data/regression_data.csv')
```
